```python
# Importing the dependencies

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
import pickle
```

```python
# Loading the csv life to a Pandas DataFrame
heart_data = pd.read_csv('./Datasets/heart.csv')
```

1 --> Defective Heart

```python
# Splitting the features and target column
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training and testing

```python
# Splitting the data into training and testing data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4, stratify=Y, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

### Logistic Regression Model Training

```python
# Logistic Regression
log_model = LogisticRegression()

# Training the Logistic Regression model with training data
log_model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
print("Logistic Regression Results:")

# Accuracy on Training Data
X_train_pred = log_model.predict(X_train)
train_accuracy = accuracy_score(X_train_pred, Y_train)
print('Accuracy on training data: ', train_accuracy)

# Accuracy on Testing Data
X_test_pred = log_model.predict(X_test)
test_accuracy = accuracy_score(X_test_pred, Y_test)
print('Accuracy on testing data: ', test_accuracy)

# Precision on Testing Data
precision = precision_score(Y_test, X_test_pred)
print('Precision:', precision)

# Recall on Testing Data
recall = recall_score(Y_test, X_test_pred)
print('Recall:', recall)

# F1 Score on Testing Data
f1 = f1_score(Y_test, X_test_pred)
print('F1 Score:', f1)
```

### Random Forest Classifier Model Training

```python
random_model = RandomForestClassifier()

random_model.fit(X_train, Y_train)
```

```python
random_test_pred = random_model.predict(X_test)
random_accuracy = accuracy_score(Y_test, random_test_pred)
random_recall = recall_score(Y_test, random_test_pred)
random_precision = precision_score(Y_test, random_test_pred)
random_f1score = f1_score(Y_test, random_test_pred)
```

```python
print('\nRandom Forest Model:')
print('Accuracy on testing data:', random_accuracy)
print('Precision:', random_precision)
print('Recall:', random_recall)
print('F1 Score:', random_f1score)
```

### Cross Validation

```python
# Perform cross-validation (e.g., 5-fold cross-validation)
cv_scores = cross_val_score(random_model, X, Y, cv=5)
```

```python
# Print the cross-validation scores
print("Cross-validation scores:", cv_scores)
print("Mean cross-validation score:", np.mean(cv_scores))
```

### Feature Analysis

```python
# Get feature importances
feature_importances = random_model.feature_importances_

# Create a DataFrame to display feature importances
feature_importance_df = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Print the feature importances
print(feature_importance_df)
```

#### Building a Predictive System

```python
input_data = (43,1,0,150,247,0,1,171,0,1.5,2,0,2)

# Change the input data to a numpyarray
input_data_numpyarray = np.asarray(input_data)

# Reshape the array as we are predicting for only one instance
input_data_reshaped = input_data_numpyarray.reshape(1,-1)

prediction = random_model.predict(input_data_reshaped)
print(prediction)

if prediction[0] == 0:
  print('The person has a healthy heart')
else:
  print('The person has some kind of a Heart Disease')
```
